# Project Introduction

Cancer has affected many people around the world. Unfortunately, we have not found a consistent, preventative cure to stop the deadly disease. In this project, we will use convolutional neural networks to classify images that will identify metastatic cancer from digital pathology scans.

Our data comes from the [Kaggle Hispatholofic Cancer Detection](https://www.kaggle.com/c/histopathologic-cancer-detection) competition. Overall, the data is split into two different sets: a training set consisting of 220,025 images and a test set consisting of 57,458. Each image has a dimension of 96 x 96 x 3. The 96 x 96 corresponds to the pixel width and height. The 3 consists of the "depth" or the color dimension of the image. In our case, we have a dimension of 3 which means our images are on the RGB color scale.

# Model Creation

We will compare three models and choose the one with the best accuracy. Note, if this wasn't a kaggle submission, we would most likely be concerned with the recall and not the accuracy. By focusing on improving recall, we can be sure we identify someone with the cancer.

- I will use a 3x3 **filter size** for each model. I want to use a smaller filter to capture the local differences in each picture. As we saw by our picture visualizations, we cannot see the differences by the naked eye. Therefore, using a smaller filter size will allow use to capture those local smaller differences. However, a 1x1 filter will create feature maps that are the same size as our input. Also, this will have a high chance of overfitting the model
- The architecture will differ in points of how many **convolutional layers** I will use. The three models will consist of 2, 4 and 5 convolutional layers. I believe, 2 layers will underfit the data. It will perform with the lowest accuracy on both the test set and dataset. However, the model with 4 convolutional layers will perform well while 5 layers will slightly overfit.
- Next, I will use **Max pooling** for the pooling layer for all models. The size of this will be 2x2. First, max pooling is used instead of average pooling becauase it is better at recognizing distringuishable features due to identifying extreme values in specific areas. Average pooling would create feature maps that don't idenify the important features, instead it would make different areas seem more similar or "smoother". Max pooling will help use identify an object no matter where it is at.
    - A **2x2** pooling is used on all models because if I use anymore there will be too much data lost since our pooling layers are essentially a generalization layer.
- **Binary Crossentropy** will be used as a loss function. Our model has a binary output. Thus using a loss function that idenitfies with a binary output is important. More specifically, we want to heavily penalize the models that have a lower probability for the correct class.  Thus, Binary Crossentropy is the perfect metric for our model.
- Again, we have a binary output classification problem. Thus, the best activation function for our output layer is the **sigmoid** function

- ** We will hypertune our **optimizer** and **learning rate** once we choose our model **

```python
#Create directories to store metrics and models
MODEL_PATH = "Models/"
if not os.path.exists(MODEL_PATH):
    os.makedirs(MODEL_PATH)

if not os.path.exists(MODEL_PATH + "Metrics"):
    os.makedirs(MODEL_PATH + "Metrics")
```

```python
#Model with 2 hidden layers
model1 = Sequential()
model1.add(Conv2D(64, (3,3), input_shape=(96,96,3)))
model1.add(Activation("relu"))
model1.add(MaxPooling2D(pool_size=(2,2)))

model1.add(Conv2D(64, (3,3)))
model1.add(Activation("relu"))
model1.add(MaxPooling2D(pool_size=(2,2)))

model1.add(Flatten())
model1.add(Dense(64))

model1.add(Dense(1))
model1.add(Activation('sigmoid'))

model1.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"]
             )
model1.fit(X_train, y_train, batch_size=128, epochs=5, validation_split=0.1)
```

```python
def plot_acc(history, axis = None, model=None, ylim=None):
    """ Plots a models validation and training accuracy"""
    axis.plot(history.history['accuracy'], label='accuracy')
    axis.plot(history.history['val_accuracy'], label='val_accuracy')
    axis.set_xlabel('Epoch')
    axis.set_ylabel('accuracy')
    axis.set_title(model)
    axis.set_ylim(ylim)
    axis.legend(loc='lower right')

def plot_val(history, axis=None, model=None):
    "Plot's a models lossat each epoch"
    axis.plot(history.history['loss'], label='loss')
    axis.plot(history.history['val_loss'], label='val_loss')
    axis.set_xlabel('Epoch')
    axis.set_ylabel('Loss')
    axis.set_title(model)
    axis.legend(loc='upper right')
```

```python
fig, ax = plt.subplots(1,3, figsize=(20,5))
plot_acc(hist1, axis=ax[0], model="Model1", ylim=[0.5, 0.65])
plot_acc(hist2, axis=ax[1], model="Model2", ylim = [0.75, 1])
plot_acc(hist3, axis=ax[2], model="Model3", ylim= [0.75, 1])
fig.tight_layout()
```

```python
fig, ax = plt.subplots(1,3, figsize=(20,5))
plot_val(hist1, axis=ax[0], model="Model1")
plot_val(hist2, axis=ax[1], model="Model2")
plot_val(hist3, axis=ax[2], model="Model3")
fig.tight_layout()
```

- Adjusting more hyperparameters

- Used less epochss
