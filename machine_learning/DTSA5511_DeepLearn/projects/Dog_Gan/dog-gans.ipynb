# Normalizing Images

Many of the picutres within the data set contain images what have dogs. Not all of the images consist of only dogs. To make sure we are training dogs, we need to crop the images so they are only modeling the characteristics of the dogs. Removing this noise will provide consistency in our model.

First, we iterate over each image until we fid one that matches the breed we are searchine for. The *Annotations* directory contain the following:
- the size of the image
- the type of dog that is in the image
- the dimensionsions that contain the dog's image

We then use these to crop the image and gain new features of our cropped image.

```python
idx = 0; breed_names = []
breed_images = np.zeros((25000,64,64,3))

#Overall, used to get cropped images that focuses dogs
for breed in breeds:
    for dog in os.listdir(ANNOTATIONS_PATH+breed):
        
        # A few images errored out so I removed from training set
        try: img = Image.open(IMAGES_PATH + dog + '.jpg') 
        except: continue
            
        #Find location of dog within image and what type
        tree = ET.parse(ANNOTATIONS_PATH + breed + '/' + dog)
        root = tree.getroot()
        objects = root.findall('object')

        #crop images so they only return images of the dogs
        for aobject in objects:
            boundaries = aobject.find('bndbox') 
            xmin = int(boundaries.find('xmin').text)
            ymin = int(boundaries.find('ymin').text)
            xmax = int(boundaries.find('xmax').text)
            ymax = int(boundaries.find('ymax').text)
            w = np.min((xmax - xmin, ymax - ymin))
            img2 = img.crop((xmin, ymin, xmin+w, ymin+w))
            img2 = img2.resize((64,64), Image.ANTIALIAS)
            breed_images[idx,:,:,:] = np.asarray(img2)
            breed_names.append(breed)
            idx += 1
```

# Model Building

- We need to create a deep neural network that is deep enough to find the important features of a dog image
- Here's why certain parameters were used
    - We use **elu** as an activation function to diminish the vanishing gradient effect
    - **Padding** is parameterized as *same* to prevent our filter from shrinking the image. We need our images to output as the same size as the input. Hence, why we also use **Reshape** layers

```python
#Build Generative Network
nn_input = Input((10000,))
layer = Dense(2048, activation='elu')(nn_input)
layer = Reshape((8,8,32))(layer)
layer = Conv2D(128, (3, 3), activation='elu', padding='same')(layer)
layer = UpSampling2D((2, 2))(layer)
layer = Conv2D(64, (3, 3), activation='elu', padding='same')(layer)
layer = UpSampling2D((2, 2))(layer)
layer = Conv2D(32, (3, 3), activation='elu', padding='same')(layer)
layer = UpSampling2D((2, 2))(layer)
output_layer = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(layer)

model_nn = Model(nn_input, output_layer)
model_nn.compile(optimizer=Adam(lr=0.003), loss='binary_crossentropy')

model_nn.summary()
```
